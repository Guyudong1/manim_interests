```python
from manim import *

class SVMExpand(Scene):
    def construct(self):
        # Step 1: 初始显示 "SVM"
        short_text = Text("SVM", font="Arial", font_size=96, color=BLUE_D)
        self.play(Write(short_text))
        self.wait(1)

        # Step 2: 展开后的完整词组
        parts = [
            ("S", "upport "),
            ("V", "ector "),
            ("M", "achine")
        ]
        expanded = VGroup()
        for cap, ext in parts:
            cap_letter = Text(cap, font="Arial", font_size=96, color=BLUE_D)
            extension = Text(ext, font="Arial", font_size=70, color=BLUE_B)
            extension.next_to(cap_letter, RIGHT, buff=0.1, aligned_edge=ORIGIN)
            extension.shift(DOWN * 0.08)
            group = VGroup(cap_letter, extension)
            expanded.add(group)

        expanded.arrize(RIGHT, buff=0.2).scale(0.9)
        expanded.move_to(short_text.get_center())

        # Step 3: Transform 动画替换
        animations = []
        fadeins = []
        for i, letter in enumerate(short_text):
            cap_letter, extension = expanded[i]
            animations.append(Transform(letter, cap_letter))
            fadeins.append(FadeIn(extension, shift=LEFT))
        self.play(*animations, *fadeins, run_time=2)

        # 关键：把 short_text 从场景中移除，避免有两层
        self.remove(short_text)
        self.add(expanded)  # 确保 expanded 在场景中
        self.wait(1)

        # Step 4: 整个 expanded (包括 S、V、M 和 support vector machine) 下移+淡出
        chinese_text = Text("支持向量机", font="Microsoft YaHei", font_size=85, color=GREEN_B)

        # 设置中文初始状态（比目标位置稍微高一点，透明）
        chinese_text.next_to(expanded, UP, buff=0.5)
        chinese_text.shift(DOWN*0.99)
        chinese_text.set_opacity(0)

        # 动画：英文整体下移 & 淡出，中文下移 & 淡入
        self.play(
            expanded.animate.shift(DOWN * 0.5).set_opacity(0),
            chinese_text.animate.shift(DOWN * 0.5).set_opacity(1),
            run_time=2
        )
        self.wait(0.5)
        self.play(chinese_text.animate.shift(DOWN * 1).set_opacity(0), run_time=1.5)
        self.wait(1)
```
